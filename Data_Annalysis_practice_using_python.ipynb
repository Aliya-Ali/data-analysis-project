```python
import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import Figure

plt.rcParams['figure.figsize'] = (12, 8)

df= pd.read_csv('/content/movies.csv')
```

```python
 df.head()
```

```python
#check the missing data
for col in df.columns:
  pct_missing = np.mean(df[col].isnull())
  print('{} - {}%'.format(col, pct_missing))
```

```python
#type of data
df.dtypes
```

```python
# Replace non-finite values with 0 (or any other value you prefer)
df['budget'] = df['budget'].replace([np.inf, -np.inf], np.nan).fillna(0)
df['gross'] = df['gross'].replace([np.inf, -np.inf], np.nan).fillna(0)
#change the data type
df['budget'] = df['budget'].astype('int64')
df['gross'] = df['gross'].astype('int64')
```

```python
df.head()
```

```python
# Create correct yearl coloumn
df['released']= df['released'].astype(str).str[:4]
```

```python
df.sort_values(by=['gross'],inplace= False, ascending=False)
df.head()
```

```python
pd.set_option('display.max_rows', None)
```

```python
df['company'] = df['company'].drop_duplicates().sort_values(ascending = False)
```

```python
df.drop_duplicates().head()
```

```python
#budget high correlation
#company high correlation
```

```python
#sactter plot with budget vs gross
plt.scatter(x=df['budget'], y=df['gross'])
plt.title('Budget vs Gross Earnings')
plt.xlabel('Gross Eaning')
plt.ylabel('Budget for film')
plt.show()
```

```python
#plot budget vs gross using seaborn

sns.regplot(x ='budget', y= 'gross', data=df, scatter_kws={"color": "red"}, line_kws={'color':"Blue"})
```

```python
for col in df.columns:
  df[col] = df[col].replace([np.inf, -np.inf], np.nan).fillna(0)
```

```python
columns=['budget', 'gross', 'runtime', 'score', 'votes', 'year']
df1 = df[columns]
df1.corr()
```

```python
#high correlation between budget and gross
```

```python
correlation_matrix = df1.corr(method = "pearson")
sns.heatmap(correlation_matrix, annot = True)
plt.title("Correlation matrix for Nemeric Features")
plt.xlabel("Movie features")
plt.ylabel("Movie features")
plt.show()
```

```python
df_numerized = df

for col_name in df_numerized.columns:
  if(df_numerized[col_name].dtype == 'object'):
    df_numerized[col_name] = df_numerized[col_name].astype('category')
    df_numerized[col_name] = df_numerized[col_name].cat.codes
```

```python
 correlation_matrix = df_numerized.corr(method = "pearson")
 sns.heatmap(correlation_matrix, annot = True)
 plt.title("Correlation matrix for Numeric Features")
 plt.xlabel("Movie features")
 plt.ylabel("Movie features")
 plt.show()
```

```python
df_numerized.corr().head()
```

```python
correlation_matrix = df_numerized.corr()
corr_pairs = correlation_matrix.unstack()
corr_pairs
```

```python
sorted_pairs = corr_pairs.sort_values()
sorted_pairs
```
